**TavernAI Launch Instructions**<br>
1. Click the TavernAI launch button.
2. Wait for the environment to load
3. After initialisation, a TavernAI link will appear
4. Click it and TavernAI will open in the new tab

**Next you need to load an AI model. For this we will use KoboldAI** [Detailed instructions (starting from p.7)](https://github.com/TavernAI/TavernAI/wiki/How-to-install)<br>
1. Go to lower where is the KoboldAI notepad section or go to [KoboldAI Colab notebook](https://colab.research.google.com/github/KoboldAI/KoboldAI-Client/blob/main/colab/GPU.ipynb#scrollTo=ewkXkyiFP2Hq)
2. Choose a model
3. Launch the section
4. Wait when the model will be loaded and copy a KoboldAI url link
5. Go to TavernAI -> open right top menu -> select "Settings" -> select KoboldAI api(usually it is selected by default) -> copy the KoboldAI url in the api url field in "Settings"
6. Click "Connect" button
